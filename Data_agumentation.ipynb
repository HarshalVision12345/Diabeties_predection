```python
from keras.preprocessing import image
from keras.src.legacy.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/dog_img.jpeg',target_size=(256,256))
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.imshow(img)
```

```python
img = image.img_to_array(img)
print(img.shape)
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    height_shift_range=0.2,
    width_shift_range=0.2,
    # fill_mode='nearest',
    fill_mode='constant'

)
```

```python
input_batch = img.reshape(1,256,256,3)
```

```python
import os

# Create the 'aug' directory if it doesn't exist
if not os.path.exists('aug'):
    os.makedirs('aug')

i = 0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='aug'):
  i += 1
  if i == 10 :
    break
```
